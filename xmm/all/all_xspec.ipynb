```python
napec = [2,2,2,2,1,1,1]
fluxes = []
fluxes_eu = []
fluxes_el = []


for i, star in enumerate(stars):
    print(star)
    
    s = Spectrum('{}_src_spectrum_bin{}.ds'.format(star, specbins[i]), backFile='{}_bkg_spectrum_grp.ds'.format(star), respFile='{}_response_grp.rmf'.format(star))#, arfFile='{}_EMOS1.arf'.format(star))
    # s.background = '{}_bkg_spectrum_grp.ds'.format(star)
    # s.response = '{}_response_grp.rmf'.format(star)
    # # s.response.arf = '{}_EMOS1.arf'.format(star)
    
    Plot('data')
    Plot.xAxis='angstrom'
    s.ignore('**-5.0 60.0-**')
    

     
    Plot.perHz=False
    Plot.area=True
    Plot('data')

    wave = Plot.x()
    rates = Plot.y()
    bins = Plot.xErr()
    error = Plot.yErr()
    

        
    
    plt.errorbar(wave, rates, xerr=bins, yerr=error, ls='none')
    AllModels.setEnergies("0.1 2.4")


    # plt.errorbar(wave, rates)
    if napec[i] == 1:
        mod = Model('apec*phabs', setPars={1:1, 2:0.4, 5:1e-4})
        mod.phabs.nH.frozen = True
        Fit.renorm()
        Fit.perform()
        Plot('data')
        folded = Plot.model()
        plt.plot(wave, folded)

    
    else:
    
        mod2 = Model('phabs*(apec+apec)', setPars={1:1e-4, 2:1,3:0.4,6:0.5, 7:0.4})
        mod2.phabs.nH.frozen = True
        Fit.renorm()
        Fit.perform()
        Plot('data')
        folded = Plot.model()
        plt.plot(wave, folded)

        
    if star == 'eguma':
        eguma_wave = np.array(wave)
        eguma_rates = np.array(rates)
        eguma_bins = np.array(bins)
        eguma_error= np.array(error)
        eguma_folded = np.array(folded)
        
    flux = AllModels.calcFlux("0.1 2.4 err")
    print(flux)
    plt.show()
    # s.flux[0]*u.erg/u.cm**2/u.s, (s.flux[0]-s.flux[1])*u.erg/u.cm**2/u.s, (s.flux[2]-s.flux[0])*u.erg/u.cm**2/u.s  
    
    fluxes.append(s.flux[0])
    fluxes_eu.append(s.flux[2]-s.flux[0])
    fluxes_el.append(s.flux[0]-s.flux[1])
    
    
    DataManager.clear(s)
    
# print(napec)   
```

```python
lstars = [-2.51, -2.51, -2.51, -1.62, -2.51, -1.93, -3.11] #estimating from spectral type
prot = [0.596, 0.668,0.563,2.018,0.246,0.931,0.337] #from pceb_bob
dstar = [39.90,28.71,42.99,176.38,220.93,189.98, 29.22] #from odette
```

```python
path = '/media/david/2tb_ext_hd/hddata/carmenes_binary/xraycat/'
xraycat = fits.open('{}J_ApJ_743_48_catalog.dat.fits'.format(path))
xraydata = xraycat[1].data
```

```python
plt.figure(figsize=(8,6))
# -12.1730
print(np.log10(fluxes))

for i, star in enumerate(stars):
    fx, fxel, fxeu = fluxes[i]*u.erg/u.cm**2/u.s, fluxes_el[i]*u.erg/u.cm**2/u.s, fluxes_eu[i]*u.erg/u.cm**2/u.s  
    d = dstar[i]*u.pc
    p = prot[i]*u.d
    lstar = (10**lstars[i])*u.Lsun
    lx = (4*np.pi)*(fx*(d.to(u.cm))**2).to(u.Lsun) #0.797 conversion factor from Wright+11
    lx_star = lx/lstar
    # print(lx_star)
    lx_star_err =  np.array(lx_star*[fxel/fx, fxeu/fx])
    # print(lx_star_err)
    if star == 'eguma':
        plt.errorbar(p, lx_star, yerr= [[lx_star_err[0]],[lx_star_err[1]]], marker='o', c='C4', markeredgecolor='k', label='EG\,UMa')
    else:
        plt.errorbar(p, lx_star, yerr= [[lx_star_err[0]],[lx_star_err[1]]], marker='o', c='C1', markeredgecolor='k', label='PCEBs')
#     if star == 'eguma':
#         fx = (10**-12.1730)*u.erg/u.cm**2/u.s
#         lx = (fx*(d.to(u.cm))**2).to(u.Lsun)
#         lx_star = lx/lstar
#         plt.errorbar(p, lx_star, yerr= [[lx_star_err[0]],[lx_star_err[1]]], marker='o', c='C2', markeredgecolor='k')
        
    
plt.scatter(xraydata['Prot'], 10**xraydata['Lx/bol'], marker='o', alpha=0.3, label='Wright et al. 2011')
plt.xscale('log')
plt.yscale('log')
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), frameon=True)

# plt.legend()
plt.xlabel('Porb (d)')
plt.ylabel('L$_{\mathrm{x}}$/L$_{\mathrm{bol}}$')
plt.tight_layout()
plt.savefig('../plots/eg_uma_Lx_prot.png', facecolor='w')
plt.savefig('../plots/eg_uma_Lx_prot.pdf', facecolor='w')
```
