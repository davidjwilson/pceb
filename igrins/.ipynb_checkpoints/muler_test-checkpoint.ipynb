```python
from muler.igrins import IGRINSSpectrum, IGRINSSpectrumList
from specutils import Spectrum1D
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
spec = IGRINSSpectrum(file='SDCH_20190318_0023.spec_a0v.fits')
```

```python
type(spec), isinstance(spec, Spectrum1D), isinstance(spec.flux, np.ndarray)
```

```python
spec = spec.normalize()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(spec.wavelength, spec.flux)
plt.ylim(0.6, 1.2)
plt.xlabel('$\lambda \;(\AA)$'); plt.ylabel('Flux');

```
