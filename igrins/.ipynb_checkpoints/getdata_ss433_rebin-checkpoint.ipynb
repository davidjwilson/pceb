```python
# Read in IGRINS SS 433 data. The data have been run through the pipeline but not
# corrected for telluric absorption or the blaze profile yet. Read in telluric star
# spectrum and divide. Average (including over orders, or truncate to prevent overlap),
# rebin?, smooth. 

# Advanced task: correct for weak residual HI absorption in the telluric A star spectra.
# Possible methods: turn A star into fits format and remove interactively in splot; 
# fit a Gaussian to the absorption and normalize in Python (interactive or feed params?).

# Read in data. Spectra are in text files with comment headers denoted by a #. The data are
# separated by date and H/K channels. Wavelengths are not sequential where orders overlap.

import numpy as np
import re
from astropy.io import ascii
import spectra_utils
import astropysics.obstools as Obs 
import astropysics.spec as Spec
from astropy.modeling import models, fitting

# SS 433 and telluric star input data for each observation:
# Nov 21 K had an extra order than the others. I deleted it by hand in the txt file.
allh = ("SS433_20140524_H_0051.txt", "SS433_20141121_H_0069.txt", "SS433_20150527_H_0080.txt", \
        "SS433_20150611_H_0085.txt", "SS433_20150730_H_0103.txt")
allk = ("SS433_20140524_K_0051.txt", "SS433_20141121_K_0069.txt", "SS433_20150527_K_0080.txt", \
        "SS433_20150611_K_0085.txt", "SS433_20150730_K_0103.txt")

tellurich = ("V1431Aql_20140524_H_0053.txt", "V1431Aql_20141121_H_0073.txt", \
             "29Vul_20150527_H_0084.txt", "32Ser_20150611_H_0081.txt", \
             "V1431Aql_20150730_H_0107.txt")
tellurick = ("V1431Aql_20140524_K_0053.txt", "V1431Aql_20141121_K_0073.txt", \
             "29Vul_20150527_K_0084.txt", "32Ser_20150611_K_0081.txt", \
             "V1431Aql_20150730_K_0107.txt")
```

```python
# Read in all spectra tables. Brute force for now.

# H data
may24h = spectra_utils.read_spec(allh[0])
may24htell = spectra_utils.read_spec(tellurich[0])
nov21h = spectra_utils.read_spec(allh[1])
nov21htell = spectra_utils.read_spec(tellurich[1])
may27h = spectra_utils.read_spec(allh[2])
may27htell = spectra_utils.read_spec(tellurich[2])
jun11h = spectra_utils.read_spec(allh[3])
jun11htell = spectra_utils.read_spec(tellurich[3])
jul30h = spectra_utils.read_spec(allh[4])
jul30htell = spectra_utils.read_spec(tellurich[4])
```

```python
# K data
may24k = spectra_utils.read_spec(allk[0])
may24ktell = spectra_utils.read_spec(tellurick[0])
nov21k = spectra_utils.read_spec(allk[1])
nov21ktell = spectra_utils.read_spec(tellurick[1])
may27k = spectra_utils.read_spec(allk[2])
may27ktell = spectra_utils.read_spec(tellurick[2])
jun11k = spectra_utils.read_spec(allk[3])
jun11ktell = spectra_utils.read_spec(tellurick[3])
jul30k = spectra_utils.read_spec(allk[4])
jul30ktell = spectra_utils.read_spec(tellurick[4])
```

```python
# Remove residual A star HI absorption. This is a small effect, about 5% deep in
# Bgamma and not apparent to the eye in the higher order lines. Also because our
# lines are so broad, this residual doesn't affect our analysis, so this correction
# is really a belt and suspenders action. Bgamma is in order 13 or 14 in K (not sure
# because of that extra order in some spectra

#screen = may24ktell['Index'] == 13
screen = (may24ktell['Wave'] > 2.164) & (may24ktell['Wave'] < 2.168)

g_init = models.Gaussian1D(amplitude=-5000., mean=2.1665, stddev=0.001) + \
models.Linear1D(slope=2.0e+6, intercept=-4.3e+6)
fit_g = fitting.LevMarLSQFitter()

# May 24:
x = may24ktell['Wave'][screen]
y = may24ktell['Counts'][screen]
g = fit_g(g_init, x, y)
#print(g)
g_new = models.Gaussian1D(amplitude=g.amplitude_0,mean=g.mean_0,stddev=g.stddev_0)
y_new = y - g_new(x)
may24ktell['Counts'][screen] = y_new
#plot(may24ktell['Wave'][screen],may24ktell['Counts'][screen])
#show()

#Nov 21:
x = nov21ktell['Wave'][screen]
y= nov21ktell['Counts'][screen]
g = fit_g(g_init,x,y)
#plot(x,y)
#plot(x,g(x))
#show()
g_new = models.Gaussian1D(amplitude=g.amplitude_0,mean=g.mean_0,stddev=g.stddev_0)
y_new = y - g_new(x)
nov21ktell['Counts'][screen] = y_new
#plot(nov21ktell['Wave'][screen],nov21ktell['Counts'][screen])
#show()

#May 24: This is so shallow we might want to skip the correction
x = may24ktell['Wave'][screen]
y= may24ktell['Counts'][screen]
g = fit_g(g_init,x,y)
#plot(x,y)
#plot(x,g(x))
#show()
g_new = models.Gaussian1D(amplitude=g.amplitude_0,mean=g.mean_0,stddev=g.stddev_0)
y_new = y - g_new(x)
may24ktell['Counts'][screen] = y_new
#plot(may24ktell['Wave'][screen],may24ktell['Counts'][screen])
#show()
#print(g)

#Jun 11:
x = jun11ktell['Wave'][screen]
y= jun11ktell['Counts'][screen]
g = fit_g(g_init,x,y)
#plot(x,y)
#plot(x,g(x))
#show()
g_new = models.Gaussian1D(amplitude=g.amplitude_0,mean=g.mean_0,stddev=g.stddev_0)
y_new = y - g_new(x)
jun11ktell['Counts'][screen] = y_new
#plot(jun11ktell['Wave'][screen],jun11ktell['Counts'][screen])
#show()

#Jul 30:
x = jul30ktell['Wave'][screen]
y= jul30ktell['Counts'][screen]
g = fit_g(g_init,x,y)
#plot(x,y)
#plot(x,g(x))
#show()
g_new = models.Gaussian1D(amplitude=g.amplitude_0,mean=g.mean_0,stddev=g.stddev_0)
y_new = y - g_new(x)
jul30ktell['Counts'][screen] = y_new
#plot(jul30ktell['Wave'][screen],jul30ktell['Counts'][screen])
#show()
```

```python
# Screen out regions of poor telluric correction
snrlim = 20.0

screen = where(may24htell['SNR'] < snrlim)
may24htell['Counts'][screen] = np.nan
screen = where(may24ktell['SNR'] < snrlim)
may24ktell['Counts'][screen] = np.nan

screen = where(nov21htell['SNR'] < snrlim)
nov21htell['Counts'][screen] = np.nan
screen = where(nov21ktell['SNR'] < snrlim)
nov21ktell['Counts'][screen] = np.nan

screen = where(may27htell['SNR'] < snrlim)
may27htell['Counts'][screen] = np.nan
screen = where(may27ktell['SNR'] < snrlim)
may27ktell['Counts'][screen] = np.nan

screen = where(jun11htell['SNR'] < snrlim)
jun11htell['Counts'][screen] = np.nan
screen = where(jun11ktell['SNR'] < snrlim)
jun11ktell['Counts'][screen] = np.nan

screen = where(jul30htell['SNR'] < snrlim)
jul30htell['Counts'][screen] = np.nan
screen = where(jul30ktell['SNR'] < snrlim)
jul30ktell['Counts'][screen] = np.nan
```

```python
# Divide by A0 star. Propagate errors:
may24hratio = may24h['Counts'] / may24htell['Counts']
may24hratioerr = sqrt(may24h['SNR']**-2 + may24htell['SNR']**-2)
may24hratioerr = may24hratio * may24hratioerr

may24kratio = may24k['Counts'] / may24ktell['Counts']
may24kratioerr = sqrt(may24k['SNR']**-2 + may24ktell['SNR']**-2)
may24kratioerr = may24kratio * may24kratioerr

nov21hratio = nov21h['Counts'] / nov21htell['Counts']
nov21hratioerr = sqrt(nov21h['SNR']**-2 + nov21htell['SNR']**-2)
nov21hratioerr = nov21hratio * nov21hratioerr

nov21kratio = nov21k['Counts'] / nov21ktell['Counts']
nov21kratioerr = sqrt(nov21k['SNR']**-2 + nov21ktell['SNR']**-2)
nov21kratioerr = nov21kratio * nov21kratioerr

may27hratio = may27h['Counts'] / may27htell['Counts']
may27hratioerr = sqrt(may27h['SNR']**-2 + may27htell['SNR']**-2)
may27hratioerr = may27hratio * may27hratioerr

may27kratio = may27k['Counts'] / may27ktell['Counts']
may27kratioerr = sqrt(may27k['SNR']**-2 + may27ktell['SNR']**-2)
may27kratioerr = may27kratio * may27kratioerr

jun11hratio = jun11h['Counts'] / jun11htell['Counts']
jun11hratioerr = sqrt(jun11h['SNR']**-2 + jun11htell['SNR']**-2)
jun11hratioerr = jun11hratio * jun11hratioerr

jun11kratio = jun11k['Counts'] / jun11ktell['Counts']
jun11kratioerr = sqrt(jun11k['SNR']**-2 + jun11ktell['SNR']**-2)
jun11kratioerr = jun11kratio * jun11kratioerr

jul30hratio = jul30h['Counts'] / jul30htell['Counts']
jul30hratioerr = sqrt(jul30h['SNR']**-2 + jul30htell['SNR']**-2)
jul30hratioerr = jul30hratio * jul30hratioerr

jul30kratio = jul30k['Counts'] / jul30ktell['Counts']
jul30kratioerr = sqrt(jul30k['SNR']**-2 + jul30ktell['SNR']**-2)
jul30kratioerr = jul30kratio * jul30kratioerr
```

```python
# New try: rebin data and average orders to remove the edge effects. 
# H data starts at 1.42972 and increments in 0.00001 micron increments.
# I believe the resl is 3 pixel sampling though we can probably bin up
# more since our lines are so broad. 

# For H: 1.4972 to 1.83470: 0.3375 micron. Let's try one resl binning
# to 0.00003, so 11250 bins. First bin centered on 1.49723

binsize = 0.00003
binstart = 1.42972
binend = 1.83470

tmpwv = np.arange(binstart+binsize/2,binend+binsize/2,binsize)
tmpflx = np.zeros(tmpwv.size)
tmpflx[:] = np.nan
tmperr = np.zeros(tmpwv.size)
tmperr[:]  = np.nan

may24h1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
nov21h1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
may27h1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jun11h1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jul30h1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
    
may24h1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
nov21h1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
may27h1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jun11h1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jul30h1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)

listh = {"may24h" : (may24h,may24hratio,may24hratioerr,may24h1d,may24h1dnorm), 
        "nov21h" : (nov21h,nov21hratio,nov21hratioerr,nov21h1d,nov21h1dnorm),
        "may27h" : (may27h,may27hratio,may27hratioerr,may27h1d,may27h1dnorm),
        "jun11h" : (jun11h,jun11hratio,jun11hratioerr,jun11h1d,jun11h1dnorm),
        "jul30h" : (jul30h,jul30hratio,jul30hratioerr,jul30h1d,jul30h1dnorm),
        }

# Scroll through each spectrum and average in rebinned, 1-D spectrum. Simple
# average introduced too much noise, so try weighted:
for key,value in listh.items():
    
    tmpflx = np.zeros(tmpwv.size)
    tmperr = np.zeros(tmpwv.size)
    
    for i in range(0,value[0]['Wave'].size -1):
        # Addl screening of bad pixels:
        if ((value[1][i] <0.05)):
                value[1][i] = np.nan
        if np.isfinite(value[1][i]):
            bin = (value[0]['Wave'][i] - binstart) / binsize
            bin = int(bin)
            tmpflx[bin] += value[1][i] / (value[2][i] * value[2][i])
            tmperr[bin] += 1 / (value[2][i] * value[2][i])
            
    value[3].x = tmpwv
    value[3].flux = tmpflx / tmperr
    value[3].err = sqrt(1.0/tmperr)

    
# For K: 1.86641 to 2.51883: 0.65242 micron. However, there are edge orders that
# I'm not sure are worth keeping. Order 2 starts at 1.90559. That would cover
# 0.61324 micron. 10221 bins.

binsize = 0.00003
binstart = 1.90559
binend = 2.51883

tmpwv = np.arange(binstart+binsize/2,binend+binsize/2,binsize)
tmpflx = np.zeros(tmpwv.size)
tmpflx[:] = np.nan
tmperr = np.zeros(tmpwv.size)
tmperr[:] = np.nan

may24k1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
nov21k1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
may27k1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jun11k1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jul30k1d = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)

may24k1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
nov21k1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
may27k1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jun11k1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
jul30k1dnorm = Spec.Spectrum(tmpwv,tmpflx,err=tmperr,name="SS 433",copy=True)
    
listk = {"may24k" : (may24k,may24kratio,may24kratioerr,may24k1d,may24k1dnorm),
         "nov21k" : (nov21k,nov21kratio,nov21kratioerr,nov21k1d,nov21k1dnorm),       
         "may27k" : (may27k,may27kratio,may27kratioerr,may27k1d,may27k1dnorm),
         "jun11k" : (jun11k,jun11kratio,jun11kratioerr,jun11k1d,jun11k1dnorm),
         "jul30k" : (jul30k,jul30kratio,jul30kratioerr,jul30k1d,jul30k1dnorm),
         }

# Scroll through each spectrum and average in rebinned, 1-D spectrum. Simple
# average introduced too much noise, so try weighted:
for key,value in listk.items():
    
    tmpflx = np.zeros(tmpwv.size)
    tmperr = np.zeros(tmpwv.size)
    tmpcnt = np.zeros(tmpwv.size)
    for i in range(0,value[0]['Wave'].size -1):
        # Addl screening of bad pixels:
        if ((value[1][i] <0.05)):
                value[1][i] = np.nan
        if np.isfinite(value[1][i]):
            bin = (value[0]['Wave'][i] - binstart) / binsize
            bin = int(bin)
            tmpflx[bin] += value[1][i] / (value[2][i] * value[2][i])
            tmperr[bin] += 1 / (value[2][i] * value[2][i])
            #tmpcnt[bin] += 1
            
    value[3].x = tmpwv
    value[3].flux = tmpflx / tmperr
    value[3].err = sqrt(1.0 / tmperr)
```
